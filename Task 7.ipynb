```python
class SmartLight:
    
    def __init__(self, light_name):
        # Initialize light name and default status
        self.light_name = light_name
        self.status = "OFF"

    def turn_on(self):
        "Method to turn the light ON"
        self.status = "ON"

    def turn_off(self):
        "Method to turn the light OFF"
        self.status = "OFF"

    def display_status(self):
        "Displays the current status of the light"
        print(f"{self.light_name} is {self.status}")


# Example Execution
if __name__ == "__main__":
    light = SmartLight("Bedroom Light")
    light.turn_on()
    light.display_status()
```

###  Problem 2: Employee ID Card System

```python
class Employee:
    def __init__(self, name, emp_id, department):
        # store employee details
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        "Prints employee ID card"
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


# Example Execution
if __name__ == "__main__":
    emp = Employee("Rahul", "EMP102", "IT")
    emp.display_id_card()
```

### Problem 3: Mobile Contact Record

```python

class Contact:
    def __init__(self, name, phone_number):
        #contact details
        self.name = name
        self.phone_number = phone_number

    def display_contact(self):
        "Displays saved contact information"
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")


# Example Execution
if __name__ == "__main__":
    contact = Contact("Anita", "9876543210")
    contact.display_contact()
```

### Problem 4: Product Price Tag Generator

```python
class Product:
    def __init__(self, product_name, price):
        # store product details
        self.product_name = product_name
        self.price = price

    def display_price_tag(self):
        "Displays price tag"
        print(f"Product: {self.product_name}")
        print(f"Price: ₹{self.price}")


# Example Execution
if __name__ == "__main__":
    product = Product("Headphones", 2499)
    product.display_price_tag()
```

### Problem 5: Movie Rating Display System

```python
class Movie:
    def __init__(self, movie_name, rating):
        # movie details
        self.movie_name = movie_name
        self.rating = rating

    def display_movie_details(self):
        """Displays movie name and rating"""
        print(f"Movie: {self.movie_name}")
        print(f"Rating: {self.rating} / 5")


# Example Execution
if __name__ == "__main__":
    movie = Movie("Inception", 4.8)
    movie.display_movie_details()
```

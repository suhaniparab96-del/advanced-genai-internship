### 1 . Given a username and password, check whether login is successfully


```python
print("Given a username and password, check whether login is successful.")

username = "admin"
password = 1234

while True:
    entered_username = input("Enter a username : ")
    entered_password = input("Enter a password : ")

    if(entered_username == username  or entered_password == password):
        print("Login Sucessful")
        break 
    else:
        print("Invalid Credential. Try Again.")
        continue
    


                       
```

### 2. Pass / Fail Analyzer

```python
print("Analyze student results.")

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0
for marks in marks:
    if marks > 50 :
        pass_count += 1
    else :
        fail_count += 1

print("Total number of pass student :" , pass_count)
print("Total number of fail student :" , fail_count)

        
```

### 4. Message Length Analyzer


```python
print("Analyze message sizes.")

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print("Message : "  , msg)
    print("Length : " , length)
```
